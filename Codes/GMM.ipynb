```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
np.random.seed(8)
X, _ = make_blobs(n_samples=100, centers=4, cluster_std=0.6, random_state=0)
```

```python
# Simulate linearly separable data and plot
fig, ax = plt.subplots(figsize = (4,4))
ax.scatter(X[:, 0], X[:, 1], c = _, cmap=plt.cm.Spectral)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
```

```python
from sklearn.mixture import GaussianMixture as GMM
```

```python
model = GMM(n_components=4)
model.fit(X)
```

```python
labels = model.predict(X)
```

```python
labels.shape
```

```python
label_prob = model.predict_proba(X)
label_prob
```

```python
np.round(label_prob, 2)
```
